```python
import json

path = "../crawling/naver-map-results.json"
save_path = "../crawling/naver-map-results-preprocessed.json"
```

```python
# JSON 파일 열기
with open(path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# 데이터 출력
print(len(data))
```

```python
# 키워드를 분류하는 함수 정의
def categorize_keywords(data):
    use_how_keywords = ['예약 후 이용', '예약 없이 이용', '포장·배달 이용']
    wait_time_keywords = ['바로 입장', '30분 이내', '10분 이내', '1시간 이상', '30분 이상', '2시간 이상']
    purpose_keywords = ['여행', '친목', '데이트', '가족모임', '나들이', '일상', '회식', '기념일', '비즈니스']
    visit_with_keywords = ['연인・배우자', '아이', '부모님', '친구', '친척・형제', '혼자', '지인・동료', '반려동물', '기타']
    
    # 결과를 저장할 딕셔너리 초기화
    categorized_data = {
        'use_how': {},
        'wait_time': {},
        'purpose': {},
        'visit_with': {}
    }
    
    # 입력 데이터를 분류
    for key, value in data.items():
        if key in use_how_keywords:
            categorized_data['use_how'][key] = value
        elif key in wait_time_keywords:
            categorized_data['wait_time'][key] = value
        elif key in purpose_keywords:
            categorized_data['purpose'][key] = value
        elif key in visit_with_keywords:
            categorized_data['visit_with'][key] = value
    
    return categorized_data
```

```python
for key in data:  # data의 모든 항목에 대해 반복
    keywords = {}
    for i in data[key]['review']:
        review = data[key]['review'][i]
        if 'visit_keywords' in review and review['visit_keywords']:  # visit_keywords가 존재하고 비어 있지 않으면
            for keyword in review['visit_keywords']:
                for k in keyword.split(','):  # keyword를 ','로 분리
                    k = k.strip()  # 공백 제거
                    if k in keywords:
                        keywords[k] += 1  # 카운트 증가
                    else:
                        keywords[k] = 1  # 초기 카운트 설정
    keywords = categorize_keywords(keywords)
    data[key]['keyword_dict'] = keywords
```

```python
with open(save_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
# JSON 파일 열기
with open(save_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# 데이터 출력
print(len(data))
```
